```python
from celery_test import task_one
from celery_test1 import task_two

# Список для хранения результатов
results = []

for i in range(100):
    if i % 2 == 0:
        result = task_one.delay(0, i)
    else:
        result = task_two.delay(0, i)

    results.append(result)

# Обработка результатов
for result in results:
    try:
        # Здесь можно использовать get(timeout=) для обработки временных ожиданий
        print(f"Result: {result.get(timeout=10)}")
    except Exception as e:
        print(f"Task failed: {e}")
```
